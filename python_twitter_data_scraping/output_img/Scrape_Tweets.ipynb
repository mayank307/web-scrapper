```python
import requests
import pandas as pd
```

```python
twitter_data = []
```

```python
payload = { 'api_key': '42cdfa5b6add0824bf2e147a1bad1080', 
            'query' : 'artificial intelligence',
            'num' : '25'
} 
response = requests.get('https://api.scraperapi.com/structured/twitter/search', params=payload)
data = response.json()
```

```python
data['organic_results'][4]['snippet']
```

```python
all_tweets = data['organic_results']
for tweet in all_tweets:
    twitter_data.append(tweet)
```

```python
df=pd.DataFrame(twitter_data)
df.to_json('tweets.json',orient='index')
print('file exported')
```
